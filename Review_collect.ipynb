# Import Required libraries 

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from time import sleep
from openpyxl import Workbook
from webdriver_manager.chrome import ChromeDriverManager
import pandas as pd

# Setup Chrome Web Driver

chrome_options = Options()
chrome_options.add_experimental_option("detach",True)

driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()),options=chrome_options)

# Sending the link for webpage

driver.get('https://www.capterra.ca/reviews/1016102/adobe-photoshop-lightroom')
sleep(5)
driver.maximize_window()
```

# Data Collecting

```python
Names = []
Bios = []
Profiles = []
Review_titles = []

name = driver.find_elements(By.XPATH,'//div[@class="h5 fw-bold mb-2"]')
bio = driver.find_elements(By.XPATH,'//div[@class="text-ash mb-2"]')
profile = driver.find_elements(By.XPATH,'//div[@class="col-12 col-md-6 col-lg-12 pt-3 pt-md-0 pt-lg-3 text-ash"]/div[1]')
review_title = driver.find_elements(By.XPATH,'//div[@class="col-lg-7"]/h3')
#next_page = driver.find_element(By.XPATH,'/html/body/main/div[4]/div/ul/li[6]/a')

for i in name:
    Names.append(i.text)

for j in bio:
    Bios.append(j.text)
    
for k in profile:
    Profiles.append(k.text)
    
for l in review_title:
    Review_titles.append(l.text)
```

# Creating Dataframe and data store in CSV file

```python
user_data = pd.DataFrame()

user_data['User Name'] = Names
user_data['User Profile'] = Profiles
user_data['User Review'] = Review_titles

user_data
```

# Data export in your local device

```python
user_data.to_csv('Sample.csv')
```
